```python
import pandas as pd
df=pd.read_csv('customer_shopping_behavior.csv')         # to fetch and read the dataset
```

```python
df.head() #to open and show the dataset
```

```python
df.info()              #to get the info of the dataset (schema kind of)
```

```python
df.describe(include='all')                              # to see the statastical summary of both numeric and categorical column
```

```python
df.isnull().sum()                  #to check how many null value 
```

```python
df['Review Rating']= df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))  #to fill the null with the median(grp by category)
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ' ,'_')                                  #to make the data consistant 
df=df.rename(columns={ 'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.columns
```

```python
#create a column age_group
labels=['Young Adult' , 'Adult' , 'Middle Aged' , 'Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)     
```

```python
df[['age','age_group']].head(10)
```

```python
#create colunms purchase_frequency_days

frequency_mapping = {
'Fortnightly': 14,
'Weekly': 7,
'Monthly': 30,
'Quarterly': 90,
'Bi-Weekly': 14,
'Annually': 365,
'Every 3 Months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied']== df['promo_code_used']).all()
```

```python
# as we can see that both columns are same so case of redundancy , so we will remove a column from our data
```

```python
df=df.drop('promo_code_used', axis=1)
```
